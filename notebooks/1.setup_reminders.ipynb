## ⚙️ Step 2 — Configure Notebook Tools

We use two tools to keep version control clean:

- **`nbdime`** — provides clean diffs and a visual merge editor for `.ipynb` files
- **`pre-commit` + `nbstripout`** — automatically strips output metadata before every `git commit`

Run the cell below to enable both for this repository.

```python
# Enable nbdime Git integration
!nbdime config-git --enable

# Install pre-commit hooks
!pre-commit install
```

You only need to run this **once per machine**. After this, `nbdime` and the pre-commit hooks are active for the lifetime of this repo clone.

---
## 🔧 Step 3 — Install the nbstripout Git Filter

This registers `nbstripout` as a **Git clean filter**, which strips notebook outputs and metadata automatically at `git add` time — before the commit hook even runs.

Without this, you may hit a double-commit issue: `pre-commit` cleans the notebook on the first attempt and the commit fails, requiring you to `git add` and retry. The filter eliminates that entirely.

```python
!nbstripout --install
```

> ⚠️ This writes to your local `.git/config` and is **not** committed to the repository. Every collaborator must run this **once** after cloning — including you, on every new clone.

## 📋 Step 4 — Collaboration Workflow

Because `.ipynb` files are JSON under the hood, standard Git tools often produce noisy, unreadable diffs. Here's how our tooling handles that:

| Situation | What happens |
|---|---|
| `git diff` | Shows only code changes — no JSON metadata noise |
| `git commit` | `nbstripout` auto-strips outputs and execution counts |
| Merge conflict on a notebook | Use the visual merge tool (see below) |

### 🔀 Handling Merge Conflicts

If you hit a conflict on a `.ipynb` file, **do not fix it manually in a text editor.** Run the cell below instead — it launches a web-based 3-pane merge editor in your browser.

```python
# Only run this if you have an active notebook merge conflict
!nbdime mergetool
```
