# 🧪 Working with Python Files in Jupyter

This notebook demonstrates the recommended workflow for this Jupyter Notebook: write logic in `.py` files under `src/`, and use notebook cells only for running and observing results.

> 🔁 `autoreload` is enabled below — any changes you make to a `.py` file will be picked up automatically without restarting the kernel.

---
## ⚙️ Step 1 — Check Environment

Before anything else, confirm that your `.env` is set up correctly. If you haven't done this yet, go to the project root and run:

```bash
cp .env.example .env
```

Your `.env` should contain:

```dotenv
PYTHONPATH=../src
```

Then **restart the Jupyter server** from inside `pipenv shell` so the `.env` is loaded into the kernel. Run the cell below to verify.

````python
# Cell 1: Universal Path Setup
from pathlib import Path
import sys

# Get the directory where THIS notebook is located.
# Path().resolve() is used instead of os.getcwd() to ensure the path
# reflects the notebook's actual location, not the kernel's launch directory.
nb_dir = Path().resolve()
print(f"Notebook is located at: {nb_dir}")

# Define the src path relative to this notebook.
# If submitting a single file, we keep src in the same folder.
src_path = nb_dir / "src"
src_path.mkdir(exist_ok=True)

# Initialize src/ as a proper Python package.
# Without __init__.py, 'import src' may behave inconsistently across environments.
init_file = src_path / "__init__.py"
if not init_file.exists():
    init_file.touch()

# Add nb_dir (not src_path) to sys.path so 'from src import ...' always works.
# Python needs to see the parent of src/ to resolve it as a package.
if str(nb_dir) not in sys.path:
    sys.path.append(str(nb_dir))

print(f"✅ Source directory set to: {src_path}")
print(f"✅ System path updated. Imports will look in: {nb_dir}")
````
