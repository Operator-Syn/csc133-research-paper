# 🧪 Working with Python Files in Jupyter

This notebook demonstrates an alternative workflow: using the `%%writefile` magic command to author source modules directly from cells into `src/`, then executing and inspecting the results inline.

> 🔁 `autoreload` will be enabled below — any changes you make to a `.py` file will be picked up automatically without restarting the kernel.

---

## ⚙️ Step 1 — Project Bootstrapping

This notebook is designed to be **self-extracting**. It will automatically generate a modular `src/` directory to store the underlying Python logic, ensuring a clean separation between research documentation and executable code.

### 🔌 Automated Path Mapping
Instead of manual `.env` configuration, the cell below programmatically maps the project structure. This ensures that:
1. The `src/` folder is created relative to this notebook's location.
2. The Python `sys.path` is updated in real-time to allow immediate imports.
3. The notebook remains **structure-agnostic** (it works whether it's in the root or a subfolder).

> **Note:** If you are running this for the first time, simply "Run All." The environment will configure itself on the fly.

---

```python
# Cell 1: Universal Path Setup
from pathlib import Path
import sys

# Get the directory where THIS notebook is located.
# Path().resolve() is used instead of os.getcwd() to ensure the path
# reflects the notebook's actual location, not the kernel's launch directory.
nb_dir = Path().resolve()
print(f"Notebook is located at: {nb_dir}")

# Define the src path relative to this notebook.
# If submitting a single file, we keep src in the same folder.
src_path = nb_dir / "src"
src_path.mkdir(exist_ok=True)

# Initialize src/ as a proper Python package.
# Without __init__.py, 'import src' may behave inconsistently across environments.
init_file = src_path / "__init__.py"
if not init_file.exists():
    init_file.touch()

# Add nb_dir (not src_path) to sys.path so 'from src import ...' always works.
# Python needs to see the parent of src/ to resolve it as a package.
if str(nb_dir) not in sys.path:
    sys.path.append(str(nb_dir))

print(f"✅ Source directory set to: {src_path}")
print(f"✅ System path updated. Imports will look in: {nb_dir}")
```

## 🧪 Step 3 — Verify Modular System (Smoke Test)

This project uses the `%%writefile` magic command to generate a modular `src/` directory. This ensures that our logic remains separated from the research documentation while allowing for a "single-file" submission.

### 📝 3.1 — Writing the Source Code
We will now generate a utility script, `src/hello_world.py`, to verify that our file paths and permissions are correctly configured.

```python
%%writefile src/hello.py
def hello_world():
    print("Hello, World from Jupyter Notebook!")


def echo(value):
    return value
```
